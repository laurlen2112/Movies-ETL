```python
#check that the adult value is boolean
kaggles_metadata['adult'].value_counts()
```

```python
# remove bad data
kaggles_metadata[~kaggles_metadata['adult'].isin(['True', 'False'])]
```

```python
# keep rows where the adult column is False, and then drop the adult column.
kaggles_metadata = kaggles_metadata[kaggles_metadata['adult'] == 'False'].drop('adult', axis = 'columns')
```

```python
kaggles_metadata['video'].value_counts()
```

```python
#creates a boolean value
kaggles_metadata['video'] == 'True'
```

```python
#assign booleans to column
kaggles_metadata['video'] = kaggles_metadata['video'] =='True'
```

```python
# change columns to number

kaggles_metadata['budget'] = kaggles_metadata['budget'].astype(int)
kaggles_metadata['id'] = pd.to_numeric(kaggles_metadata['id'], errors = 'raise')
kaggles_metadata['popularity'] = pd.to_numeric(kaggles_metadata['popularity'], errors = 'raise')
```

```python
#convert release date to date
kaggles_metadata['realease_date'] = pd.to_datetime(kaggles_metadata['release_date'])
```
